# Parsing Historical Data using gemini multimodal

```python
import os
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np


import PIL.Image
```

```python
image = PIL.Image.open('UP_census_51.png')
```

```python
f, ax = plt.subplots(1, 1, figsize=(15, 15))
ax.imshow(image)
```

```python
from google import genai
```

```python
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
```

```python
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=["""
              This is a digital image of a census table from a historical census of British India.
              The rows are district names, and contain counts broken down by Total, Rural, and Urban.
              Generate a plaintext comma-delimited table that contains the information in the table.
              """,
    image]
)
```

```python
print(response.text)
```
